```python
# Dependencies
import numpy as np
import csv
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import sqlite3
```

```python
df = pd.read_csv('postgres_upload_file.csv')
cols = df.columns
```

```python
connection = sqlite3.connect('test_dbase') 
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS yelp_test1(
business_id text PRIMARY KEY,
city text,
state text,
postal_code integer,
stars float,
review_count integer,
diversity integer)
""")
                     
connection.commit()
```

```python
with open('postgres_upload_file.csv','r') as fin:
    dr = csv.DictReader(fin)
    to_db = [(i['business_id'], i['city'], i['state'], i['postal_code'], i['stars'], i['review_count'], i['diversity']) for i in dr]     

cursor.executemany("INSERT INTO yelp_test1 (business_id,city,state,postal_code,stars,review_count,diversity) VALUES (?, ?, ?, ?, ?, ?, ?);", to_db)
connection.commit()
connection.close()
```
