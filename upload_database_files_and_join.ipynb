# Join Two Tables in AWS RDS with SQL Alchemy

```python
# use sql alchemy to join the two tables in aws rds

sql = '''
SELECT stars.postal_code, sum.pizza, stars.pizza_stars
FROM zip_cat_stars_avg_final AS stars
JOIN zip_categories_sum_final AS sum ON stars.postal_code = sum.postal_code
WHERE stars.postal_code BETWEEN 1 AND 50000
'''

q1 = pd.read_sql(sql,con=connection)
connection.close()
q1.head(10)
```
