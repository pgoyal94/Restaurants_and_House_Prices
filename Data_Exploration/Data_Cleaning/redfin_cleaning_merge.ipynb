```python
import pandas as pd
from functools import reduce
```

```python
#load data
df = pd.read_csv('Resources/redfin_raw.tsv000', sep='\t')
```

```python
#list of columns in dataframe
df.columns.to_list()
```

```python
df.head()
```

```python
df.shape
```

```python
#datatypes
df.dtypes
```

```python
#uravel period_end and split into mulitple columns
df[["year", "month","date"]]=df["period_end"].str.split('-', expand=True)
```

```python
df.head()
```

```python
#drop unneeded columns
df = df.drop(columns=['period_begin',
 'period_end',
 'period_duration',
 'region_type',
 'region_type_id',
 'table_id',
 'is_seasonally_adjusted','state_code','property_type_id','median_sale_price_mom',
 'median_sale_price_yoy',
 'median_list_price',
 'median_list_price_mom',
 'median_list_price_yoy',
 'median_ppsf',
 'median_ppsf_mom',
 'median_ppsf_yoy',
 'median_list_ppsf',
 'median_list_ppsf_mom',
 'median_list_ppsf_yoy','homes_sold_mom',
 'homes_sold_yoy',
 'pending_sales',
 'pending_sales_mom',
 'pending_sales_yoy',
 'new_listings',
 'new_listings_mom',
 'new_listings_yoy',
 'inventory',
 'inventory_mom',
 'inventory_yoy',
 'months_of_supply',
 'months_of_supply_mom',
 'months_of_supply_yoy',
 'median_dom',
 'median_dom_mom',
 'median_dom_yoy',
 'avg_sale_to_list',
 'avg_sale_to_list_mom',
 'avg_sale_to_list_yoy',
 'sold_above_list',
 'sold_above_list_mom',
 'sold_above_list_yoy',
 'price_drops',
 'price_drops_mom',
 'price_drops_yoy',
 'off_market_in_two_weeks',
 'off_market_in_two_weeks_mom',
 'off_market_in_two_weeks_yoy',
 'parent_metro_region',
 'parent_metro_region_metro_code',
 'last_updated'], axis=1)
```

```python
#filter by only house prices for 2021
df = df[df["year"]=="2021"]
```

```python
df.head()
```

```python
#unravel the region column into type and postal code
df[["type", "postal_code"]]=df["region"].str.split(":", expand=True)
df.head()
```

```python
df.shape
```

```python
#unravel categorical column:property_type into multiple columns
```

```python
df.property_type.value_counts()
```

```python
df_redfin = pd.get_dummies(df, columns=['property_type'], prefix="")
```

```python
df_redfin.head()
```

```python
#create a dataframe that has per postal_code the mean house price, count of property types
price = df_redfin.groupby(["postal_code"]).mean()["median_sale_price"].round(0)
residential = df_redfin.groupby(["postal_code"]).sum()["_All Residential"]
multifamily = df_redfin.groupby(["postal_code"]).sum()["_Multi-Family (2-4 Unit)"]
condo_co_op = df_redfin.groupby(["postal_code"]).sum()["_Condo/Co-op"]
single_family = df_redfin.groupby(["postal_code"]).sum()["_Single Family Residential"]
townhouse = df_redfin.groupby(["postal_code"]).sum()["_Townhouse"]

price_df = pd.DataFrame({"2021": price,
                        "All_Residential":residential,
                        "Multi_Family_homes": multifamily,
                        "Condo/Co-op": condo_co_op,
                        "Single_Family_Homes":single_family,
                        "Towmhomes":townhouse})
price_df.head()
```

```python
price_df.reset_index()
```

```python
df_redfin.columns
```

```python
df_redfin = df_redfin.drop(columns=['median_sale_price', 'homes_sold', 'year',
       'month', 'date', 'type', '_All Residential',
       '_Condo/Co-op', '_Multi-Family (2-4 Unit)',
       '_Single Family Residential', '_Townhouse', 'region', 'city'])
```

```python
#drop duplicate postal_code
df_redfin = df_redfin.drop_duplicates(subset=["postal_code"])
```

```python
df_redfin.shape
```

```python
# Merge dataframes
frames = [df_redfin, price_df]
redfin_df = reduce(lambda left,right: pd.merge(left,right,on='postal_code'), frames)
print(redfin_df.shape)
redfin_df.head()
```

```python
redfin_df.state.value_counts()
```

```python
redfin_df["postal_code"].nunique()
```
