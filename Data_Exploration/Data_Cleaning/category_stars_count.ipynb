```python
import pandas as pd
```

```python
path = "expanded_restaurants.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.shape
```

```python
stars_df = df[["postal_code", "stars"]]
```

```python
stars_df.head()
```

```python
stars_df["stars"].value_counts()
```

```python
df_stars = pd.get_dummies(stars_df, columns=["stars"])
```

```python
df_stars.head()
```

```python
df_stars.shape
```

```python
df_stars = df_stars.groupby(["postal_code"]).sum()
```

```python
df_stars.head()
```

```python
df_stars.astype('int32').dtypes
```

```python
df_stars["total_stars"]=df_stars.iloc[0:].sum(axis=1)
```

```python
df_stars.head()
```

```python
df_stars["ratio_stars_1.0"] = df_stars["stars_1.0"]/df_stars["total_stars"]
df_stars["ratio_stars_1.5"] = df_stars["stars_1.5"]/df_stars["total_stars"]
df_stars["ratio_stars_2.0"] = df_stars["stars_2.0"]/df_stars["total_stars"]
df_stars["ratio_stars_2.5"] = df_stars["stars_2.5"]/df_stars["total_stars"]
df_stars["ratio_stars_3.0"] = df_stars["stars_3.0"]/df_stars["total_stars"]
df_stars["ratio_stars_3.5"] = df_stars["stars_3.5"]/df_stars["total_stars"]
df_stars["ratio_stars_4.0"] = df_stars["stars_4.0"]/df_stars["total_stars"]
df_stars["ratio_stars_4.5"] = df_stars["stars_4.5"]/df_stars["total_stars"]
df_stars["ratio_stars_5.0"] = df_stars["stars_5.0"]/df_stars["total_stars"]
df_stars.head()
```

```python
# Reset index 
df_stars= df_stars.reset_index()
df_stars.head()
```

```python
df_stars.to_csv("stars_count.csv")
```
