```python
import pandas as pd
```

```python
#Read in Census data for NYC
df = pd.read_csv('Resources/NYC_census.csv', index_col=False)
```

```python
df.head()
```

```python
df = df.set_index('Label (Grouping)')
```

```python
#Flip rows and columns to match other data sets to join
df_transposed = df.transpose().reset_index()
df_transposed
```

```python
df_transposed = df_transposed.rename(columns={"index": "identifier"})
df_transposed
```

```python
#Split out the data type
df_transposed[['identifier','data_type']] = df_transposed['identifier'].str.split('!!', expand=True)
```

```python
#Split out the postal code
df_transposed[['identifier','postal_code']] = df_transposed['identifier'].str.split(' ', expand=True)
df_transposed
```

```python
#Keep only the 'Estimate' data type, now we have a single row per zip code
census_df = df_transposed[df_transposed.data_type == 'Estimate']
```

```python
census_df = census_df.dropna(axis=1)
```

```python
census_df
```

```python
#drop unnecessary columns
census_df = census_df.drop(columns=['identifier','data_type'])
census_df
```

```python
census_df.columns
```

```python
#pick only relevant columns
census_df2 = census_df.iloc[:,[89,0,3,17,36,37,38,43,51,56,57,85]]
```

```python
census_df2
```

```python
census_df2.dtypes
```

```python
#Remove commas as they are stopping conversions from str to float
census_df2 = census_df2.apply(lambda x: x.str.replace(',', ''))
census_df
```

```python
#find na values
census_df2.isna().sum()
```

```python
#drop na values
census_df2 = census_df2.dropna()
census_df2
```

```python
#Remove '-' and spaces as they are stopping conversions from str to float
census_df2 = census_df2.apply(lambda x: x.str.replace('-', ''))
census_df2 = census_df2.apply(lambda x: x.str.replace(' ', ''))
census_df2
```

```python
len(census_df2.columns)
```

```python
#Write DF to csv to clean fill blanks with 0s in CSV as code wasn't working
census_df2.to_csv('Resources/census_df2.csv')
```
